## Tarefa 3
### Quais os termos usados para Aspirina (consulta JSON/MeSH em Python)?

Buscando o código da Aspirina:

```python
url_template = "https://id.nlm.nih.gov/mesh/lookup/descriptor?label={concept}&match={match}&limit={limit}"

concept = "Aspirin"
match = "exact"
limit = "10"

concept = concept.replace(" ", "%20")

url = url_template.replace("{concept}", concept)\
                  .replace("{match}", match)\
                  .replace("{limit}", limit)

data = requests.get(url)
json_asp = data.json()
print(json.dumps(json_asp, indent=3))
```

Isolando o código:

```python
uri_mesh = json_asp[0]["resource"]
print("uri do recurso: " + uri_mesh)
bar = str.rindex(uri_mesh, "/")
print("código do recurso: " + uri_mesh[bar+1:])
```

Buscando os detalhes da aspirina a partir do código:

```python
url_template = "https://id.nlm.nih.gov/mesh/lookup/details?descriptor={code}"

code = uri_mesh[bar+1:]

url = url_template.replace("{code}", code)

data = requests.get(url)
json_aspd = data.json()
print(json.dumps(json_aspd, indent=3))
```

Isolando os termos para aspirina:

```python
terms = []
for d in json_aspd['terms']:
  terms.append(d['label'])

print(terms)
```

## Tarefa 4
### Escolha algum lugar no OpenStreetMap e elabore a mesma sequência de três chamadas de serviços para o seu conceito.

Pesquisa pelo Central Park:

```python
url_template = "http://nominatim.openstreetmap.org/search?format={format}&q={query}"
format = "json"
query = "Central Park"

url = url_template.replace("{format}", format)\
                  .replace("{query}", query)

data = requests.get(url)
park_ref = data.json()
print(json.dumps(park_ref, indent=3))
```

Extraindo detalhes a partir do OSM ID:

```python
url_template = "https://api.openstreetmap.org/api/0.6/{element_type}/{id}.{format}"

element_type = "way" # types: node, way, or relation
id = str(park_ref[0]["osm_id"])
format = "json"

url = url_template.replace("{element_type}", element_type)\
                  .replace("{id}", id)\
                  .replace("{format}", format)

data = requests.get(url)
park_way = data.json()
print(json.dumps(park_way, indent=3))
```

Buscando no DBPedia:

```python
url_template = "http://dbpedia.org/data/{concept}.{format}"
concept = park_way["elements"][0]["tags"]["wikipedia"].replace(" ", "_")\
                                                        .replace("en:", "")
format = "jsod"

url = url_template.replace("{concept}", concept)\
                  .replace("{format}", format)
print(url)
data = requests.get(url)
park_dbpedia = data.json()
print(json.dumps(park_dbpedia, indent=3))
```
