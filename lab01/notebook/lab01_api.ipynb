## Banco da UNICEF
Possui uma API REST acessada por HTTP, e uma ferramenta online para auxiliar na construção da URL para uma dada requisição, que pode ser encontrada [aqui](https://sdmx.data.unicef.org/webservice/data.html).

Foi construída a URL para acessar alguns dados globais de mortalidade infantil, em formato CSV.

```python
import pandas as pd
import io
import requests
url = "https://sdmx.data.unicef.org/ws/public/sdmxapi/rest/data/UNICEF,CME,1.0/WORLD.CME_MRY5T14+CME_SB+CME_TMY0T4._T._T?format=csv&labels=both"
data = requests.get(url).content
csv = pd.read_csv(io.StringIO(data.decode("utf-8")))
print(csv)
```
